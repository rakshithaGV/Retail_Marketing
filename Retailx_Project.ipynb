```python
fact_orders['order_date']=pd.to_datetime(fact_orders['order_date']) # conver orderdate into date format
```

```python
fact_orders['revenue'] = fact_orders['quantity'] * fact_orders['unit_price'] * (1 - fact_orders['discount']/100)
fact_orders.head()
```

```python
#1. Revenue By Category

#Group by category
revenue_by_category = fact_orders.groupby('category').agg(
    total_revenue=('revenue','sum'),
    total_orders=('order_id','count'),
    avg_order_value=('revenue','mean')
).reset_index()

#Sort by total revenue (descending)
revenue_by_category = revenue_by_category.sort_values(by='total_revenue',ascending=False).round(2)
print(revenue_by_category)
```

```python
#export as output_revenue_by_category
revenue_by_category.to_csv(r"D:\My Work\Project\output_revenue_by_category.csv", index=False)
```

```python
#2. Extracting year-month from order_date
fact_orders['year_month'] = fact_orders['order_date'].dt.to_period('M')
fact_orders.head()
```

```python
# Group by year_month and compute metrics
monthly_sales = fact_orders.groupby('year_month').agg(
    total_revenue=('revenue', 'sum'),
    total_orders=('order_id', 'count')
).reset_index()
print(monthly_sales)
```

```python
# Calculate Month-over-Month (MoM) Growth percentage

monthly_sales['mom_growth_pct'] = (
    monthly_sales['total_revenue'].pct_change() * 100 ).round(2).fillna(0) # pct_change() → calculates % change from previous row & * 100 → converts to percentage

print(monthly_sales)
```

```python
# Sort chronologically
monthly_sales = monthly_sales.sort_values('year_month')
print(monthly_sales)
```

```python
monthly_sales.to_csv(r"D:\My Work\Project\output_monthly_sales.csv", index=False) # Export monthly sales data
```

```python
#3. Counting distinct orders per customer

orders_per_customer = fact_orders.groupby('customer_id')['order_id'].nunique().reset_index() # Group data by customer_id and count distinct order_id
orders_per_customer.rename(columns={'order_id': 'order_count'}, inplace=True) # Rename column for clarity
print(orders_per_customer)
```

```python
# Identify one-time buyers

one_time_customers = orders_per_customer[orders_per_customer['order_count'] == 1] # Filter customers who placed only 1 order
print(one_time_customers) # Every customer likely has more than 1 order hence getting nothing in output
```

```python
# Calculate percentage of one-time buyers

total_customers = orders_per_customer.shape[0] # Total number of customers
one_time_count = one_time_customers.shape[0] # Number of one-time buyers
one_time_percentage = (one_time_count / total_customers) * 100 # Calculating percentage of one time buyers
print(f"Percentage of one-time buyers: {one_time_percentage:.2f}%") #Printing the result clearly

#output analysis
# Meaning: Every row = one unique order, No duplicate orders
# Now if each customer has: 2 or more orders
# Then: No one has exactly 1 order
# So: one_time_customers = EMPTY
```

```python
# Exporting one-time customers list
one_time_customers.to_csv(r"D:\My Work\Project\output_churn_customers.csv", index=False)
```

```python
#4. Business Question 4 – Average order value by Region

# Aggregate data by region
aov_by_region = fact_orders.groupby('region').agg(
    total_revenue=('revenue', 'sum'),       # Total revenue per region
    total_orders=('order_id', 'count'),     # Number of orders per region
    avg_order_value=('revenue', 'mean')     # Average order value (AOV)
).reset_index().round(2) # reset index and Round numeric values to 2 decimal places
```

```python
# Sort regions by highest average order value
aov_by_region = aov_by_region.sort_values(
    by='avg_order_value', ascending=False)
print(aov_by_region)
```

```python
# exporting output to CSV file
aov_by_region.to_csv(r"D:\My Work\Project\output_aov_by_region.csv", index=False)
```

```python
#5. Returns & Cancellations

# Group by order status and count number of orders
status_summary = fact_orders.groupby('status').agg(
    total_orders=('order_id', 'count'),     # Count of orders per status
    total_revenue=('revenue', 'sum')        # Total revenue per status
).reset_index()
```

```python
# Calculating percentage of total orders

total_orders_all = fact_orders['order_id'].count() # Get total number of orders
status_summary['order_percentage'] = (status_summary['total_orders'] / total_orders_all) * 100 # Calculate percentage contribution of each status
status_summary['order_percentage']= status_summary['order_percentage'].round(2)
print(status_summary)
```

```python
# Calculating revenue loss (Returned & Cancelled)

loss_fact_orders = fact_orders[fact_orders['status'].isin(['Returned', 'Cancelled'])] # Filter only Returned and Cancelled orders
revenue_lost = loss_fact_orders['revenue'].sum() # Calculate total revenue lost
print(f"Total revenue lost due to returns and cancellations: {revenue_lost:.2f}") # Print revenue loss clearlywith 2 decimal factor
```

```python
# export summary to CSV
status_summary.to_csv(r"D:\My Work\Project\output_order_status.csv", index=False)
```
